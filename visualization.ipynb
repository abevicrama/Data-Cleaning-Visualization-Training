```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Ice Cream Ratings.csv')
df = df.set_index('Date')
df
```

```python
df.plot(kind = 'bar', subplots = True, title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Score')
```

```python
df.plot(kind = 'bar',stacked = True)
```

```python
df['Texture Rating'].plot(kind = 'bar',stacked = True)
```

```python
df.plot.barh(stacked = True)
```

```python
df.plot.scatter(x ='Texture Rating', y ='Flavor Rating',s = 100, c= 'red')
```

```python
df.plot.hist(bins = 20)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,4))
```

```python
plt.style.use('seaborn-v0_8')
```

```python
df.plot.pie(y = 'Flavor Rating', figsize = (10,6))
```
